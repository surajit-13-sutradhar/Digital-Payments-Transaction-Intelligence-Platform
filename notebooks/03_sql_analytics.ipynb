```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../warehouse/payments_warehouse.db")

def run(query):
    return pd.read_sql(query, conn)
```

```python
def run_file(filepath):
    with open(filepath, "r") as f:
        query = f.read()
    return run(query)
```
